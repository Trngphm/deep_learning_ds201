```python
import json

files = ["train_word.json", "dev_word.json", "test_word.json"]
input_dir = "./data/phoner/"
output_dir = "./data/phoner/"

for file in files:
    input_file = input_dir + file
    output_file = output_dir + file.replace(".json", "_fixed.json")
    
    with open(input_file, "r", encoding="utf-8") as f:
        lines = f.read().strip().split("\n")
    
    # Chuyển từng dòng JSON object thành list
    objects = [json.loads(line) for line in lines]
    
    # Ghi ra file mới
    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(objects, f, ensure_ascii=False, indent=2)
    
    print(f"Đã fix file: {file} -> {output_file}")
```
